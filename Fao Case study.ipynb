```python
new_temp = temp.drop(['Domain Code', 'Element Code', 'Year Code', 'Months Code','Flag Description' ], axis=1)
new_temp.head()
```

```python
new_temp.mean()
```

```python
new_temp.value_counts('Value')
```

```python
new_temp.value_counts('Months')
```

```python
#using grooup by
new_temp.groupby(["Area"]).mean()
```

```python
area_or_month_base_grouping = new_temp.groupby(["Area", "Months"]).mean()
```

```python
area_or_month_base_grouping
```

```python
new_temp[new_temp ['Months'] == 'January'].mean()
```

```python
new_temp[new_temp ['Months'] == 'January'].groupby(["Area", "Months"]).mean()
```

```python
new_temp[new_temp ['Months'] == 'January'].groupby(["Area", "Months"]).sum()
```

```python
plt.figure(figsize=(12,8))
sns.lineplot(x="Year", y="Value", hue="Area",data=new_temp).set(title="Consolidated overview of Climate change across SAARC region during 1961-2021")
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(data=new_temp, x="Area", y="Value").set(title="Distribution of Temperature Change in past 50 years by Area (Country)")
```
